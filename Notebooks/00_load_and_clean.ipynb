```python
from pathlib import Path
import pandas as pd
import numpy as np

PROJECT_DIR = Path("..").resolve()

RAW_DIR  = PROJECT_DIR / "Data" / "Raw"
PROC_DIR = PROJECT_DIR / "Data" / "Processed"

assert RAW_DIR.exists(), f"RAW_DIR not found: {RAW_DIR}"
PROC_DIR.mkdir(parents=True, exist_ok=True)

print("PROJECT_DIR:", PROJECT_DIR)
print("RAW_DIR    :", RAW_DIR)
print("PROC_DIR   :", PROC_DIR)
```

```python
dictionary = pd.read_csv(RAW_DIR / "dictionary.csv")

drv10 = pd.read_csv(RAW_DIR / "master_panel_drv10.csv")
drv15 = pd.read_csv(RAW_DIR / "master_panel_drv15.csv")
drv30 = pd.read_csv(RAW_DIR / "master_panel_drv30.csv")

scoring = pd.read_csv(RAW_DIR / "scoring.csv")

print("Loaded shapes:")
print("dictionary:", dictionary.shape)
print("drv10     :", drv10.shape)
print("drv15     :", drv15.shape)
print("drv30     :", drv30.shape)
print("scoring   :", scoring.shape)
```

```python
def standardize_columns(df):
    df = df.copy()
    df.columns = (
        df.columns.astype(str)
        .str.strip()
        .str.lower()
        .str.replace(r"\s+", "_", regex=True)
    )
    return df

dictionary = standardize_columns(dictionary)
drv10 = standardize_columns(drv10)
drv15 = standardize_columns(drv15)
drv30 = standardize_columns(drv30)
scoring = standardize_columns(scoring)
```

```python
def drop_exact_duplicates(df, name):
    before = len(df)
    df = df.drop_duplicates()
    after = len(df)
    print(f"{name}: removed {before - after} duplicate rows")
    return df

dictionary = drop_exact_duplicates(dictionary, "dictionary")
drv10 = drop_exact_duplicates(drv10, "drv10")
drv15 = drop_exact_duplicates(drv15, "drv15")
drv30 = drop_exact_duplicates(drv30, "drv30")
scoring = drop_exact_duplicates(scoring, "scoring")
```

```python
MISSING_MARKERS = {"", "NA", "N/A", "na", "null", "None"}

def normalize_missing(df):
    df = df.copy()
    for c in df.columns:
        if df[c].dtype == "object":
            df[c] = df[c].replace(list(MISSING_MARKERS), np.nan)
    return df

dictionary = normalize_missing(dictionary)
drv10 = normalize_missing(drv10)
drv15 = normalize_missing(drv15)
drv30 = normalize_missing(drv30)
scoring = normalize_missing(scoring)
```

```python
dictionary.to_csv(PROC_DIR / "dictionary_clean.csv", index=False)
drv10.to_csv(PROC_DIR / "drv10_clean.csv", index=False)
drv15.to_csv(PROC_DIR / "drv15_clean.csv", index=False)
drv30.to_csv(PROC_DIR / "drv30_clean.csv", index=False)
scoring.to_csv(PROC_DIR / "scoring_clean.csv", index=False)

print("Clean files saved to:", PROC_DIR)
```
